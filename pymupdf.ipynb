# 표(테이블) 추출

```python
# 전체 페이지에서 표(테이블) 추출하기
from pprint import pprint

page = doc[0] # get the 1st page of the document
tabs = page.find_tables() # locate and extract any tables on page
print(f"{len(tabs.tables)} found on {page}") # display number of found tables

if tabs.tables:  # at least one table found?
   # pprint(tabs[0].extract())  # print content of first table
   pprint(tabs[0].extract())  # print content of first table
```

```python
def list_to_markdown_table(nested_list):
    markdown_table = []

    # 첫 번째 행을 헤더로 추가
    header = nested_list[0]
    header_row = '| ' + ' | '.join([str(cell).replace('\n', '<br>') if cell is not None and cell != '' else 'nan' for cell in header]) + ' |'
    markdown_table.append(header_row)

    # 헤더 밑에 구분선 추가
    separator = '| ' + ' | '.join(['---' for cell in header]) + ' |'
    markdown_table.append(separator)

    # 나머지 행들을 데이터로 추가
    for row in nested_list[1:]:
        markdown_row = '| ' + ' | '.join([str(cell).replace('\n', '<br>') if cell is not None and cell != '' else 'nan' for cell in row]) + ' |'
        markdown_table.append(markdown_row)

    return '\n'.join(markdown_table)
```

```python
markdown_table = list_to_markdown_table(tabs[0].extract())
```

```python
# 전체 페이지에서 테이블 추출해 파일로 저장하기
for page_index in range(len(doc)): # iterate over pdf pages
    page = doc[page_index] # get the page

    tabs = page.find_tables() # locate and extract any tables on page

    # print the number of tables found on the page
    if tabs.tables:
        print(f"Found {len(tabs.tables)} tables on page {page_index}")
    else:
        print("No tabless found on page", page_index)

    for table_index, tab in enumerate(tabs.tables, start=1): # enumerate the table list
        file_path = "./data/pymupdf/extracted_tables/page_%s-table_%s.md" % (page_index, table_index)
        with open(file_path, "w", encoding="utf-8") as file:
            # tab.extract() 메소드를 통해 리스트 형태로 출력된 표의 내용을 마크다운 형식으로 변환해서 파일로 저장.
            file.write(list_to_markdown_table(tab.extract()))
            file.close()
```

# 이미지 추출

```python
# 전체 페이지에서 이미지 가져오기
for page_index in range(len(doc)):
    page = doc[page_index]  # 현재 페이지 객체 가져오기
    image_list = page.get_images()  # 현재 페이지에 포함된 이미지 모두 가져오기
    print('현재 페이지: ', page_index, ', 이미지 갯수: ', len(image_list), ', 이미지 리스트: ', image_list)
```

```python
# 전체 페이지에서 이미지 추출해 파일로 저장하기
for page_index in range(len(doc)): # iterate over pdf pages
    page = doc[page_index] # get the page
    image_list = page.get_images()

    # print the number of images found on the page
    if image_list:
        print(f"Found {len(image_list)} images on page {page_index}")
    else:
        print("No images found on page", page_index)

    for image_index, img in enumerate(image_list, start=1): # enumerate the image list
        xref = img[0] # get the XREF of the image
        pix = pymupdf.Pixmap(doc, xref) # create a Pixmap

        if pix.n - pix.alpha > 3: # CMYK: convert to RGB first
            pix = pymupdf.Pixmap(pymupdf.csRGB, pix)

        pix.save("./data/pymupdf/extracted_img/page_%s-image_%s.png" % (page_index, image_index)) # save the image as png
        pix = None
```

# LayoutAnalysis and Rendering

```python
import fitz
import matplotlib.patches as patches
import matplotlib.pyplot as plt
from PIL import Image


def plot_pdf_with_boxes(pdf_page, segments):
    pix = pdf_page.get_pixmap()
    pil_image = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)

    fig, ax = plt.subplots(1, figsize=(10, 10))
    ax.imshow(pil_image)
    categories = set()
    category_to_color = {
        "Title": "orchid",
        "Image": "forestgreen",
        "Table": "tomato",
    }
    for segment in segments:
        points = segment["coordinates"]["points"]
        layout_width = segment["coordinates"]["layout_width"]
        layout_height = segment["coordinates"]["layout_height"]
        scaled_points = [
            (x * pix.width / layout_width, y * pix.height / layout_height)
            for x, y in points
        ]
        box_color = category_to_color.get(segment["category"], "deepskyblue")
        categories.add(segment["category"])
        rect = patches.Polygon(
            scaled_points, linewidth=1, edgecolor=box_color, facecolor="none"
        )
        ax.add_patch(rect)

    # Make legend
    legend_handles = [patches.Patch(color="deepskyblue", label="Text")]
    for category in ["Title", "Image", "Table"]:
        if category in categories:
            legend_handles.append(
                patches.Patch(color=category_to_color[category], label=category)
            )
    ax.axis("off")
    ax.legend(handles=legend_handles, loc="upper right")
    plt.tight_layout()
    plt.show()


def render_page(doc_list: list, page_number: int, print_text=True) -> None:
    pdf_page = fitz.open(file_path).load_page(page_number - 1)
    page_docs = [
        doc for doc in doc_list if doc.metadata.get("page_number") == page_number
    ]
    segments = [doc.metadata for doc in page_docs]
    plot_pdf_with_boxes(pdf_page, segments)
    if print_text:
        for doc in page_docs:
            print(f"{doc.page_content}\n")
```
