```python
# Main quest 2
# 아이펠 코어/ 8기/ 강대식

# [메인 퀘스트 2번: 아이펠 통합 관리 프로그램 - 2점]
# 아이펠 캠퍼스에서는 그루들에게 학번을 부여하고 퀘스트 점수를 관리하고 있습니다. 이제부터 그루들의 정보와
# 학번과 점수를 관리할 수 있는 Aiffel 클래스를 만들어보도록 하겠습니다.

# Q1. Aiffel 클래스 만들기
# 생성자에는 이름, 입학연도, 기수, 과정명의 인자를 받습니다. 이를 통해 학번을 생성하고 학번 정보로 8기 코어
# 그루 명단을 관리해봅시다.
# 학번은 7자리로 이루어져 있습니다. : YYNNCXX
# ● 학번 생성 규칙:
# ○ YY: 연도 마지막 두자리 (2024 -> 24)
# ○ NN: 기수 (온라인 8기는 31기)
# ○ C: 과정 종류 (리서치 = 1, 코어 = 2)
# ○ XX: 가나다순 번호

# Q2. 점수 관리 메서드
# 다음과 같은 기능이 추가되어야 합니다.
# ● 퀘스트 점수 부여
# ● 퀘스트 점수 차감 (페널티 부여)
# ● 현재 총점 출력

# Q3. 그루 명단 관리하기
# 다음과 같은 기능을 가진 프로그램을 만들어 봅시다.
# ● 전체 명단과 정보를 보는 기능
# ● 퀘스트 점수를 기준으로 내림차순으로 정렬해서 보는 기능
# ● 특정 인물 정보를 삭제하는 기능

# Q4. 그룹핑 기능 추가하기
# 그루 명단을 기준으로 2인, 3인, 4인 조를 랜덤하게 구성하는 기능을 추가해봅시다.
```

```python
# [메인 퀘스트 2번: 아이펠 통합 관리 프로그램

# 생성자에는 이름, 입학연도, 기수, 과정명의 인자를 받습니다. 이를 통해 학번을 생성하고 학번 정보로 8기 코어
# 그루 명단을 관리해봅시다.

class TotalManagementSystem:

    def __init__(self, name, enr_year, grade, course):
        self.name = name
        self.enr_year = enr_year
        self.grade = grade
        self.course = course

        # 초기 퀘스트 점수는 0으로 초기화
        self.quest_score = 0


    # 학번을 생성
    def gen_student_id(self, alphabetical_index):
        # YY: 연도 마지막 두자리 (2024 -> 24)
        enrol_year = str(self.enr_year)[-2:]
        # NN: 기수 (온라인 8기는 31기)
        grade_code = f"{str(self.grade)}"
        # C: 과정 종류 (리서치 = 1, 코어 = 2)
        course_code = '1' if self.course == '리서치' else '2'
        # XX: 가나다순 번호
        alphabetical_index_code = f"{alphabetical_index:02}"
        
        # 학번 생성
        student_id = enrol_year + grade_code + course_code + alphabetical_index_code

        return student_id
    
    
    # 퀘스트 점수 부여 (points)
    def add_quest_score(self, points):
        self.quest_score += points


    # 퀘스트 점수 차감 (페널티 부여) (penalty)
    def deduct_quest_score(self, penalty):
        if self.quest_score >= penalty:
            self.quest_score -= penalty
        else:
            print("퀘스트 점수가 부족합니다. 패널티를 적용할 수 없습니다.")
    

    # 현재 총점 = 현재 가지고 있는 점수 - 차감점수
    def total_score(self):
        total = self.quest_score
        return total
    

    # def display_current_info(self):
        # print("이름", ":", self.name, "학번", ":",self.student_id, "점수", ":", self.quest_score, "패널티", ":", self.penalty, "총점", ":", self.total_score)
    


    
```
