```python
# Main Quest 1
# [메인 퀘스트 1번: 은행계좌 만들기 - 3점]
# Q1. Account 클래스 : 은행에 가서 계좌를 개설하면 은행이름, 예금주, 계좌번호, 잔액이 설정됩니다. Account 클래스를 생성한 후
#     생성자(hint: 매직매서드..!!)를 구현해보세요. 생성자에서는 예금주와 초기 잔액만 입력 받습니다. 은행이름은 SC은행으로 계좌번호는
#     3자리-2자리-6자리 형태로 랜덤하게 생성됩니다. (은행이름: SC은행, 계좌번호: 111-11-111111)
# Q2. 클래스 변수: 클래스 변수를 사용해서 Account 클래스로부터 생성된 계좌 객체의 개수를 저장하세요.
# Q3. 클래스 변수 출력 : Account 클래스로부터 생성된 계좌의 개수를 출력하는 get_account_num() 메서드를 추가하세요.
# Q4. 입금 메서드 : Account 클래스에 입금을 위한 deposit 메서드를 추가하세요. 입금은 최소 1원 이상만 가능합니다.
# Q5. 출금 메서드 : Account 클래스에 출금을 위한 withdraw 메서드를 추가하세요. 출금은 계좌의 잔고 이상으로 출금할 수는 없습니다.
# Q6. 정보 출력 메서드 : Account 인스턴스에 저장된 정보를 출력하는 display_info() 메서드를 추가하세요. 잔고는 세자리마다 쉼표를
#     출력하세요.    (은행이름: SC은행, 예금주: 파이썬, 계좌번호: 111-11-111111, 잔고: 10,000원)
# Q7. 이자 지급하기 : 입금 횟수가 5회가 될 때 잔고를 기준으로 1%의 이자가 잔고에 추가되도록 코드를 변경해보세요.
# Q8. 여러 객체 생성 : Account 클래스로부터 3개 이상 인스턴스를 생성하고 생성된 인스턴스를 리스트에 저장해보세요.
# Q9. 객체 순회 반복문을 통해 리스트에 있는 객체를 순회하면서 잔고가 100만원 이상인 고객의 정보만 출력하세요.
# Q10. 입금과 출금 내역이 기록되도록 코드를 업데이트 하세요.
# (입금 내역과 출금 내역을 출력하는 deposit_history와 withdraw_history 메서드를 추가하세요.)
```

```python
# [메인 퀘스트 1번: 은행계좌 만들기]
import random

# Q1. Account class 생성
class Account:
    # 생성된 계좌 객체의 개수 초기화
    account_count = 0

    # 예금주, 초기 잔액만 입력 받음
    def __init__(self, name, balance):
        # 입금 횟수 초기화
        self.deposit_count = 0
        self.with_history = []
        self.dep_history = []

        self.name = name
        self.balance = balance
        self.bank = "SC은행"
        # 3-2-6 (계좌번호: 111-11-111111)
        num1 = random.randint(0, 999)
        num2 = random.randint(0, 99)
        num3 = random.randint(0,999999)
        # 숫자가 작아 빈자리가 생겼을 때 0을 채운다. ex) 001-01-000001
        num1 = str(num1).zfill(3)
        num2 = str(num2).zfill(2)
        num3 = str(num3).zfill(6)
        self.account_no = num1 + '-' + num2 + '-' + num3

        # Q2. 클래스 변수를 사용해서 Account 클래스로부터 생성된 계좌 객체의 개수를 저장
        Account.account_count += 1

    # Account 클래스로부터 생성된 계좌의 개수를 출력하는 get_account_num() 메서드 추가
    @classmethod
    def get_account_num(cls):
        print(cls.account_count)

    # Account 클래스에 입금을 위한 deposit 메서드를 추가. 입금은 최소 1원 이상만 가능
    def deposit(self, dep_amount):
        if dep_amount >= 1:
            self.dep_history.append(dep_amount)
            self.balance += dep_amount

            # 이자 지급하기: 입금 횟수가 5회가 될 때 잔고를 기준으로 1%의 이자가 잔고에 추가되도록 코드 변경
            self.deposit_count += 1
            if self.deposit_count % 5 == 0:
                self.balance = (self.balance * 1.01)

    # Account 클래스에 출금을 위한 withdraw 메서드를 추가. 출금은 계좌의 잔고 이상으로 출금 x
    def withdraw(self, with_amount):
        if self.balance > with_amount:
            # 출금 내역 저장
            self.with_history.append(with_amount) 
            self.balance -= with_amount

    # Account 인스턴스에 저장된 정보를 출력하는 display_info() 메서드를 추가. 잔고는 세자리마다 쉼표를 출력
    # (은행이름: SC은행, 예금주: 파이썬, 계좌번호: 111-11-111111, 잔고: 10,000원)
    def display_info(self):
        print("1. 은행이름: ", self.bank) 
        print("2. 예 금 주: ", self.name)
        print("3. 계좌번호: ", self.account_no)
        print("4. 계좌잔고: ", self.balance, "원")


    # 입금과 출금 내역이 기록되도록 코드를 업데이트.
    # (입금 내역과 출금 내역을 출력하는 deposit_history와 withdraw_history 메서드 추가)
    def withdraw_history(self):
        print("[ 입금 내역 ]")
        for with_amount in self.with_history:
            count = 0
            count += 1
            print(count, "입금: ", with_amount)
            return with_amount

    def deposit_history(self):
        count = 0
        print("[ 출금 내역 ]")
        for dep_amount in self.dep_history:
            count += 1
            print(count, "출금: ", dep_amount)
```
